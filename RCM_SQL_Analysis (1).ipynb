```python
#What percentage of claims belong to each outcome?
query = """
SELECT outcome,
       COUNT(*) AS Total_Claims,
       ROUND(COUNT(*) * 100.0 /
       (SELECT COUNT(*) FROM claims),2) AS Percentage
FROM claims
GROUP BY outcome
ORDER BY Percentage DESC;
"""

pd.read_sql(query, conn)
```

```python
#Paid Claims with Prior Authorization
query = """
SELECT prior_auth_obtained,
       COUNT(*) AS Paid_Claims
FROM claims
WHERE outcome='paid'
GROUP BY prior_auth_obtained;
"""

pd.read_sql(query, conn)
```

```python
#Pending Claims with Prior Author
query = """
SELECT prior_auth_obtained,
       COUNT(*) AS Pending_Claims
FROM claims
WHERE outcome='pending'
GROUP BY prior_auth_obtained;
"""

pd.read_sql(query, conn)
```

```python
#Revenue by Prior Authorization Status
query = """
SELECT prior_auth_obtained,
       ROUND(SUM(claim_amount_usd),2) AS Revenue
FROM claims
GROUP BY prior_auth_obtained;
"""

pd.read_sql(query, conn)
```

```python
#Percentage of Claims with Prior Authorization
query = """
SELECT
ROUND(
COUNT(CASE WHEN prior_auth_obtained=1 THEN 1 END)
*100.0/COUNT(*),2)
AS Prior_Auth_Percentage
FROM claims;
"""

pd.read_sql(query, conn)
```
